# La Probabilidad

La probabilidad es una rama de las matemáticas que estudia los fenómenos aleatorios o no deterministas. Se encarga de medir la certeza o incertidumbre de que un evento ocurra.

## Conceptos básicos

- **Experimento aleatorio**: Proceso que produce un resultado no predecible con certeza.
- **Espacio muestral (Ω)**: Conjunto de todos los posibles resultados de un experimento.
- **Evento**: Subconjunto del espacio muestral.
- **Probabilidad de un evento**: Valor entre 0 y 1 que representa la posibilidad de ocurrencia.

## Propiedades fundamentales

1. La probabilidad de cualquier evento está entre 0 y 1: 0 ≤ P(A) ≤ 1
2. La probabilidad del espacio muestral completo es 1: P(Ω) = 1
3. La probabilidad del evento vacío es 0: P(∅) = 0
4. Para eventos mutuamente excluyentes: P(A ∪ B) = P(A) + P(B)

## Tipos de probabilidad

- **Probabilidad clásica**: Basada en resultados equiprobables.
- **Probabilidad frecuentista**: Basada en la frecuencia relativa de ocurrencia.
- **Probabilidad subjetiva**: Basada en creencias personales sobre la posibilidad de un evento.

La probabilidad es fundamental en estadística, ciencia de datos, física, economía, y muchos otros campos que analizan incertidumbre y toman decisiones basadas en datos.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definimos un conjunto de colores (ahora con nombres en inglés)
colores_nombres = ["rojo", "azul", "verde", "amarillo", "naranja"]
colores_plt = ["red", "blue", "green", "yellow", "orange"]  # Colores válidos para matplotlib

# Calculamos la probabilidad de sacar cada color (asumiendo equiprobabilidad)
n_colores = len(colores_nombres)
probabilidad_por_color = 1/n_colores

# Mostramos la probabilidad para cada color
print(f"Probabilidad de sacar cada color: {probabilidad_por_color:.2f} o {probabilidad_por_color*100:.0f}%")

# Simulación de 1000 selecciones aleatorias
np.random.seed(42)  # Para reproducibilidad
selecciones = np.random.choice(colores_nombres, size=1000)

# Contamos las frecuencias
frecuencias = {color: np.sum(selecciones == color)/len(selecciones) for color in colores_nombres}

# Visualizamos los resultados
plt.figure(figsize=(10, 6))
plt.bar(frecuencias.keys(), frecuencias.values(), color=colores_plt)
plt.axhline(y=probabilidad_por_color, color='black', linestyle='--', 
            label=f'Probabilidad teórica: {probabilidad_por_color:.2f}')
plt.xlabel('Color')
plt.ylabel('Frecuencia relativa')
plt.title('Simulación de selección aleatoria de colores')
plt.legend()
plt.ylim(0, 0.3)
plt.show()

print("\nExplicación:")
print("- En este ejemplo tenemos 5 colores distintos.")
print("- La probabilidad teórica de seleccionar cualquier color es 1/5 = 0.2 (20%).")
print("- Hemos simulado 1000 selecciones aleatorias para ver cómo se comporta la frecuencia relativa.")
print("- Según la ley de los grandes números, a medida que aumentamos el número de pruebas,")
print("  la frecuencia relativa se aproxima a la probabilidad teórica.")
````

# Probabilidad Condicional

La probabilidad condicional mide la posibilidad de que ocurra un evento, sabiendo que otro evento ya ha ocurrido. Esta relación se denota como P(B|A), que se lee "probabilidad de B dado A".

## Definición formal

La probabilidad condicional de un evento B dado un evento A se calcula mediante la fórmula:

$$P(B|A) = \frac{P(A \cap B)}{P(A)}$$

donde:
- P(A ∩ B) es la probabilidad conjunta de A y B
- P(A) es la probabilidad del evento A (debe ser mayor que cero)

## Propiedades importantes

- Si los eventos A y B son independientes, entonces P(B|A) = P(B)
- P(A|A) = 1, la probabilidad de A dado que A ha ocurrido es 1
- La probabilidad condicional cumple con los axiomas básicos de probabilidad: está entre 0 y 1, y la suma de las probabilidades condicionales de todos los posibles resultados es 1

## Aplicaciones

La probabilidad condicional es fundamental en:
- Teorema de Bayes
- Cadenas de Markov
- Análisis de riesgos
- Diagnósticos médicos
- Sistemas de recomendación
- Filtros de spam

Entender la probabilidad condicional es esencial para analizar correctamente situaciones donde eventos no son independientes, permitiendo tomar decisiones más informadas basadas en información previa.

# Ejemplo de Probabilidad Condicional

Consideremos un ejemplo clásico para ilustrar la probabilidad condicional:

## Ejemplo: Diagnóstico médico

Supongamos que estamos analizando una prueba médica para una enfermedad poco común:

- La prevalencia de la enfermedad en la población es del 1% (P(E) = 0.01)
- La prueba tiene una sensibilidad del 95%, es decir, detecta correctamente al 95% de los enfermos (P(+|E) = 0.95)
- La prueba tiene una especificidad del 90%, es decir, identifica correctamente al 90% de las personas sanas (P(-|no E) = 0.90)

**Pregunta**: Si una persona obtiene un resultado positivo, ¿cuál es la probabilidad de que realmente tenga la enfermedad?

**Solución**: Buscamos P(E|+), que podemos calcular usando el Teorema de Bayes:

$$P(E|+) = \frac{P(+|E) \times P(E)}{P(+)}$$

Donde P(+) = P(+|E)×P(E) + P(+|no E)×P(no E)

Sustituyendo:
- P(+|E) = 0.95 (sensibilidad)
- P(E) = 0.01 (prevalencia)
- P(+|no E) = 0.10 (1 - especificidad)
- P(no E) = 0.99

$$P(E|+) = \frac{0.95 \times 0.01}{0.95 \times 0.01 + 0.10 \times 0.99} = \frac{0.0095}{0.0095 + 0.099} = \frac{0.0095}{0.1085} \approx 0.088$$

Por tanto, a pesar de obtener un resultado positivo, la probabilidad de tener realmente la enfermedad es de solo 8.8%, lo que puede resultar sorprendente. Este fenómeno, conocido como la "falacia del fiscal", ilustra por qué la probabilidad condicional es tan importante en la interpretación correcta de los resultados de pruebas diagnósticas.

# Probabilidad Conjunta

La probabilidad conjunta es una medida que indica la probabilidad de que ocurran simultáneamente dos o más eventos. Se denota como P(A ∩ B) o simplemente P(A, B), y se lee "probabilidad de A y B".

## Definición formal

La probabilidad conjunta de dos eventos A y B se define como la probabilidad de que ambos eventos ocurran al mismo tiempo:

$$P(A \cap B) = P(A, B)$$

## Relación con la probabilidad condicional

La probabilidad conjunta está relacionada con la probabilidad condicional mediante la fórmula:

$$P(A \cap B) = P(A|B) \times P(B) = P(B|A) \times P(A)$$

## Propiedades importantes

- 0 ≤ P(A ∩ B) ≤ min(P(A), P(B))
- Si A y B son eventos independientes, entonces P(A ∩ B) = P(A) × P(B)
- Si A y B son mutuamente excluyentes, entonces P(A ∩ B) = 0

## Aplicaciones

La probabilidad conjunta es fundamental en:
- Análisis de riesgos correlacionados
- Modelos de dependencia estadística
- Análisis multivariado
- Teoría de decisiones
- Procesamiento de datos en inteligencia artificial

Comprender la probabilidad conjunta permite analizar cómo diferentes eventos se relacionan entre sí, lo que es esencial para modelar fenómenos complejos donde múltiples variables interactúan simultáneamente.

# Ejemplo de Probabilidad Conjunta

Consideremos un ejemplo para ilustrar la probabilidad conjunta:

## Ejemplo: Lanzamiento de dos dados

Supongamos que lanzamos dos dados justos de seis caras y queremos analizar diferentes probabilidades:

- El espacio muestral contiene 36 resultados posibles (6×6)
- Cada resultado individual tiene probabilidad 1/36

### Algunas probabilidades conjuntas:

- P(primer dado = 3, segundo dado = 4) = 1/36
- P(suma = 7) = 6/36 = 1/6 (hay 6 combinaciones que suman 7)
- P(ambos dados muestran el mismo número) = 6/36 = 1/6 (hay 6 casos)

### Verificación de independencia:

Consideremos los eventos:
- A: "El primer dado muestra un 3"
- B: "El segundo dado muestra un 4"

Comprobemos si son independientes:
- P(A) = 1/6
- P(B) = 1/6
- P(A ∩ B) = 1/36 = (1/6) × (1/6) = P(A) × P(B)

Como P(A ∩ B) = P(A) × P(B), confirmamos que los resultados de ambos dados son eventos independientes.

Este ejemplo ilustra cómo la probabilidad conjunta nos permite analizar la ocurrencia simultánea de eventos y evaluar su independencia, conceptos fundamentales en el análisis probabilístico.

# Naive Bayes (Naïve Bayes)

## Idea central (intuición)
Naive Bayes es una familia de clasificadores probabilísticos basada en el Teorema de Bayes. La hipótesis "naive" (ingenua) es asumir que las características son condicionalmente independientes entre sí dado la clase. Esa suposición simplifica enormemente el cálculo de la probabilidad condicional de la observación completa:

P(C | x) ∝ P(C) · P(x | C) ≈ P(C) · ∏_{i} P(x_i | C)

Donde:
- C es una clase (etiqueta).
- x = (x_1, x_2, ..., x_n) es el vector de características.
- P(C) es la probabilidad previa (prior) de la clase.
- P(x_i | C) es la probabilidad condicional de la característica i dado C.

## Fórmula (Bayes y regla de decisión)
Aplicando Bayes y la suposición de independencia condicional:

P(C | x) = P(C) P(x | C) / P(x)  
Con independencia condicional:
P(x | C) = ∏_{i=1}^{n} P(x_i | C)

Para clasificar se usa la regla MAP (Maximum a Posteriori):

ŷ = argmax_{C} P(C) ∏_{i} P(x_i | C)

En práctica se trabaja en log-espacio para evitar underflow:

ŷ = argmax_{C} [ log P(C) + Σ_{i} log P(x_i | C) ]

## Estimación de parámetros (entrenamiento)
Dependiendo del tipo de datos hay variantes de Naive Bayes:

1. Gaussian Naive Bayes (features continuas)
    - Modelo: P(x_i | C) = Normal(x_i | μ_{i,C}, σ_{i,C}^2)
    - Estimadores (MLE):
      μ_{i,C} = promedio de x_i en la clase C
      σ_{i,C}^2 = varianza de x_i en la clase C

2. Multinomial Naive Bayes (conteos, texto con bag-of-words)
    - Modelo para conteos: P(x | C) basado en multinomial sobre el vocabulario.
    - Estimador (con suavizado de Laplace α):
      P(word_j | C) = (count_{j,C} + α) / (Σ_{k} count_{k,C} + αK)
      donde K = número de palabras/vocabulario.

3. Bernoulli Naive Bayes (características binarias)
    - Modelo de presencia/ausencia: P(x_i=1 | C) estimada por proporción de documentos de C que contienen la característica i.
    - Suavizado de Laplace también recomendado:
      P(x_i=1 | C) = (N_{i,C} + α) / (N_C + 2α)

Suavizado (Laplace): añade α (usualmente 1) para evitar probabilidades cero cuando aparece una característica no vista en entrenamiento.

## Ejemplo numérico sencillo (paso a paso)
Problema: clasificar entre Spam (S) y Ham (H) con dos características binarias f1,f2.

Datos estimados:
- Prior: P(S)=0.4, P(H)=0.6
- P(f1=1 | S)=0.75, P(f2=1 | S)=0.5
- P(f1=1 | H)=0.2,  P(f2=1 | H)=0.1

Queremos clasificar x = (f1=1, f2=0).

Calcular P(x | S) ≈ P(f1=1|S)·P(f2=0|S) = 0.75 · (1 - 0.5) = 0.75 · 0.5 = 0.375  
Posterior no normalizada: P(S)·P(x|S) = 0.4 · 0.375 = 0.15

Calcular P(x | H) = 0.2 · (1 - 0.1) = 0.2 · 0.9 = 0.18  
Posterior no normalizada: 0.6 · 0.18 = 0.108

Comparando: 0.15 > 0.108 → elegir S (Spam).  
Si se quiere la probabilidad normalizada, dividir entre la suma 0.15+0.108.

Este ejemplo muestra cómo multiplicar condicionales simples y combinar con el prior.

## Implementación práctica (breve)
En scikit-learn hay implementaciones listas:
- GaussianNB para datos continuos.
- MultinomialNB para conteos (texto).
- BernoulliNB para binarias.

Ejemplo de uso (markdown con código de ejemplo):
```python

```

Consejos prácticos:
- Convertir texto a conteos (CountVectorizer / TfidfVectorizer) para MultinomialNB.
- Escalar o transformar características continuas si usan GaussianNB (no obligatorio).
- Usar log-probabilidades en implementaciones propias.
- Aplicar suavizado (α>0) para evitar ceros.

## Ventajas
- Muy rápido de entrenar y predecir.
- Requiere pocos datos para estimar parámetros.
- Funciona sorprendentemente bien en muchas tareas de texto (spam, clasificación de documentos).
- Modelo interpretable (se pueden inspeccionar las probabilidades condicionadas).

## Limitaciones
- La suposición de independencia condicional rara vez es cierta en datos reales. Si las características están fuertemente correlacionadas, el rendimiento puede degradarse.
- Las probabilidades predichas pueden no estar bien calibradas (se usan más como ordenamiento).
- Para variables continuas GaussianNB asume distribución normal; si no se cumple, conviene transformar o usar otro modelo.

## Consejos para mejorar resultados
- Eliminar o combinar características muy correlacionadas.
- Aplicar selección de características (chi-cuadrado, mutual information) para texto.
- Normalizar/features engineering según la variante usada.
- Si necesitas probabilidades calibradas, aplicar calibradores (CalibratedClassifierCV).
- Usar Laplace (o α pequeño) para evitar ceros con vocabularios grandes.

## Cuándo usar Naive Bayes
- Como baseline rápido para clasificación.
- En problemas de texto (document classification, spam, análisis de sentimientos).
- Cuando se necesita un modelo simple, interpretable y eficiente.
- En escenarios con muchas características y pocos datos, puede ser robusto si la independencia no es extremadamente violada.

Resumen: Naive Bayes convierte el Teorema de Bayes en un clasificador práctico asumiendo independencia condicional. Es simple, eficiente y a menudo sorprendentemente efectivo, especialmente en texto, pero hay que vigilar la validez de la suposición de independencia y usar suavizado y transformaciones adecuadas según el tipo de dato.

* Deriva del Teorema de Bayes
* EI Teorema de Bayes describe la probabilidad condicional de un evento, basado en el
conocimiento previo de las condiciones que podrían estar relacionadas con el mismo.
* En problemas de clasificación, nos permite determinar la probabilidad condicional de que un
determinado ejemplo pertenezca a una clase particular dadas sus características
* Siguiendo el teorema de Bayes, se debería calcular las relaciones de probabilidad entre todas
las características del conjunto de datos, 10 que provoca un elevado coste computacional
* EI teorema de Naive Bayes simplifica este proceso suponiendo independencia condicional
entre las características 

# Teorema de Bayes

$$
P(c|x) = \frac{P(x|c)P(c)}{P(x)}
$$

+ C es la clase o la categorizacion de una variable
+ X son las caracteristicas de los valores

$P(c|x) \rightarrow$ El termino de la igualdad quiere saber que probabilidad hay de que para una categoria C los valores de los conjuntos den esa categoria

$P(x) \rightarrow$ El parametro es la probabilidad de obtener esa categoria dentro de las categorias que dan caracteristica a los valores

$P(x|c) \rightarrow$ El parametro aqui es donde mas consume recursos, donde tiene que buscar entre el conjunto de datos descompone la probabilidad de ser una caracteristica de una clase en concreto como $P(x_1|c)*P(x_2|c,x_1)*P(x_3|c,x_1,x_2) \rightarrow P(x_1|c)*P(x_2|c)*P(x_1)*P(x_3|c)*P(x_1)*P(x_2)$ 

$$
P(c|x) = \frac{P(x|c)P(c)}{P(x)}
\rightarrow
P(c|x_1,x_2,...x_n) = \frac{P(x_1,x_2,...x_n|c)P(c)}{P(x_1,x_2,...x_n)}
$$

El algoritmo Naybe Bayes implementa un concepto llamado __Independencia Condicional__. De esta manera el algoritmo divide los terminos x en terminos independientes.

$$
P(x_i|c,x_j) = P(x_i|c)
$$

Y con ello se tiene un cambio en la ecuacion:

$$
P(c|x_1,x_2,...x_n) = \frac{P(x_1|c)P(x_2|c)...P(x_n|c)P(c)}{P(x_1|c)P(x_2|c)...P(x_n|c)}
$$

**Ejemplo**

| x1 | x2   | c    |
|----|------|------|
| ?! | HTML | SPAM |
| 2 | 5 | 1 |
| 2 | 1 | 0 |
| 2 | 6 | 1 |
| 4 | 5 | 1 |
| 1 | 0 | 0 |

Hay una cantidad de 5 emails en este ejemplo. Y entonces yo quiero ver la probabilidad de colocar $(x_1,x_2) \rightarrow (1, 5)$.

$P(1|1,5) = \frac{P(1|1)P(5|1)P(1)}{P(1)P(5)} = \frac{0 * \frac{2}{3} * \frac{3}{5}}{\frac{1}{5} * \frac{2}{5}} = 0 \rightarrow 0,41$

Para que esto no ocurra hay algo que se llama correccion de *Laplace* y ese 0 cambia a 1 y por tanto cambia el valor. Y lo mismo se puede hacer con los diferentes valores que hay con la categoria y las caracteristicas.


### Tipos de algoritmos Naive Bayes
+ Multinomial Naive Bayes: EI clasificador Naive Bayes multinomial es adecuado para la clasificación con características discretas (por ejemplo, recuentos de palabras para la clasificación de texto).
+ Bernoulli Naive Bayes: A1 igual que MultinomialNB, este clasificador es adecuado para datos que mientras MultinomialNB funciona con recuentos de ocurrencia, BernoulliNB está diseñado para funciones binarias/booleanas. Funciona bien para detección de Spam
+ Gaussian Naive Bayes: Es apropiado para datos continuos que se distribuyen conforme a o Gaussiana.
