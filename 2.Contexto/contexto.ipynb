# CONTEXTO DE LA IA
La ia realmente esta cambiando la manera en que estamos actuando día a día, respuestas, preguntas, mejoras, revoluciones, identificaciones, clasificaciones.
Ahora la IA tiene un uso mayor a lo que hacia desde un inicio, pero ahora el aprendizaje de maquina, el aprendizaje profundo y demas tipos que evolucionan el actuar de las maquinas nos ha llevado a poder mejorar tanto a los ajentes de computadora que permita ejecutar tareas, automatizar procesos y hasta simplemente hacer acciones repetitivas como tambien complejas.

Años 50-70: El Nacimiento y los Primeros Pasos

* Concepto: En 1956, en la conferencia de Dartmouth, se acuñó el término "Inteligencia Artificial".
* Enfoque: La investigación se centraba en la resolución de problemas y el razonamiento simbólico. Se crearon programas que podían resolver problemas de álgebra, probar teoremas lógicos y jugar a las damas.
* Limitaciones: La falta de poder computacional y de datos limitó el progreso, llevando al llamado "primer invierno de la IA".

Años 80: Los Sistemas Expertos

* Auge: La IA experimentó un resurgimiento comercial con los "sistemas expertos".
* Funcionamiento: Estos programas imitaban el conocimiento y la capacidad de decisión de un experto humano en un dominio muy específico (por ejemplo, diagnóstico médico o configuración de sistemas informáticos).
* Problemas: Eran costosos de crear, difíciles de mantener y muy frágiles fuera de su área de especialización, lo que condujo a un segundo "invierno de la IA".

Años 90-2000: La Era del Aprendizaje Automático (Machine Learning)

* Cambio de Paradigma: En lugar de programar reglas explícitamente, los sistemas comenzaron a aprender patrones a partir de grandes cantidades de datos.
* Impulso: El aumento de la potencia de los ordenadores y la explosión de datos gracias a Internet fueron clave.
* Aplicaciones: Algoritmos como las redes neuronales, los árboles de decisión y las máquinas de vectores de soporte (SVM) se hicieron populares para tareas como el filtrado de spam, los sistemas de recomendación y el reconocimiento de escritura.

Años 2010-Presente: La Revolución del Aprendizaje Profundo (Deep Learning)

* Avance Clave: El uso de redes neuronales mucho más grandes y complejas ("profundas") y el desarrollo de hardware especializado (como las GPUs) permitieron avances espectaculares.
* Hitos: En 2012, un modelo de aprendizaje profundo (AlexNet) ganó de forma contundente una competición de reconocimiento de imágenes, demostrando el poder de este enfoque.
Capacidades Actuales: Esto ha llevado a la IA que conocemos hoy, con capacidades avanzadas en:
    * Visión por Computadora: Reconocimiento facial, conducción autónoma.
    * Procesamiento del Lenguaje Natural (PLN): Traductores automáticos, asistentes de voz como Siri o Alexa, y modelos de lenguaje avanzados como los que impulsan a los chatbots.
    * IA Generativa: Creación de texto, imágenes, música y código completamente nuevos.

## CLASIFICANDO IA

- Artificial Superintelligence (ASI):
    * Más allá de una inteligencia humana que lleva la vista al futuro pero por ahora 2025 no hay alguna que este en este tipo.
- Artificial General Intelligence (AGI):
    * Sistema capaz de ejecutar tareas como un humano, pero en 2025 no hay aun sistemas que pueden clasificarse aca.
- Artificial Narrow Intelligence (ANI):
    * Se puede mejorar el rendimiento pero no demasiado y todos los modelos realmente estan en este tipo.

## CATEGORIAS IA

- Self Aware: Consientes de su estado interno, no existen 2025.

- Theory of Mind: Incorpora la capacidad de entender pensamientos al estilo de C-3PO de star wars no esta en 2025.

- Limited Memory: Incorporan memoria de fragmentos pasados, donde varias aqui se ubican.

- Purely Reactive: No memoria ni experiencias, es muy responsibo como lo de los videojuegos que reaccionan en el momento.
